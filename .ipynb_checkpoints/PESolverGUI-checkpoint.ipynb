```python
from PyQt6.QtWidgets import (QApplication, QMainWindow, QWidget, 
                             QVBoxLayout, QPushButton, QSlider, 
                                QLabel, QSize)
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg
import matplotlib.pyplot as plt

class PESolverGUI(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("PE Solver")
        self.setFixedSize(QSize(400, 300))
        
        # Create central widget
        central = QWidget()
        self.setCentralWidget(central)
        layout = QVBoxLayout(central)
        
        # Add controls
        self.freq_slider = QSlider()
        self.freq_slider.setRange(10, 500)
        self.freq_slider.setSingleStep(5)
        layout.addWidget(QLabel("Frequency [Hz]"))
        layout.addWidget(self.freq_slider)
        
        # Add matplotlib canvas
        self.figure = plt.figure()
        self.canvas = FigureCanvasQTAgg(self.figure)
        layout.addWidget(self.canvas)
        
        # Add run button
        run_btn = QPushButton("Run Simulation")
        run_btn.clicked.connect(self.run_simulation)
        layout.addWidget(run_btn)
    
    def run_simulation(self):
        freq = self.freq_slider.value()
        # Run solver and update plot
        self.figure.clear()
        ax = self.figure.add_subplot(111)
        # ... plotting code ...
        self.canvas.draw()

if __name__ == "__main__":
    app = QApplication([])
    window = PESolverGUI()
    window.show()
    app.exec()
```
